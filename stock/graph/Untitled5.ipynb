```
import matplotlib.pyplot as plt
import numpy as np
```

```
x = [1, 2, 3]
y = [5, 7, 4]

x2 = [1, 2, 3]
y2 = [10, 14, 12]

plt.plot(x, y, label='first line')
plt.plot(x2, y2, label='second line')
plt.xlabel('Flot number')
plt.ylabel('Important var')
plt.legend()
plt.title('Graph')
plt.show()
```

```
x = [2, 4, 6, 8, 10]
y = [6, 7, 8, 2, 4]

x2 = [1, 3, 5, 9, 10]
y2 = [7, 8, 2, 4, 2]
plt.bar(x, y, label='Bars1', color='r')
plt.bar(x2, y2, label='Bars2')
plt.legend()
plt.show()
```

```
population_ages = [22, 55, 62, 45, 21, 22, 34, 42, 42, 4, 99, 102, 110, 120, 121, 122, 130, 111, 115, 112, 80, 75, 65, 54, 44, 43, 42]

# ids = [x for x in range(len(population_ages))]

bins = [x for x in range(0, max(population_ages)+1, 10)]

plt.hist(population_ages, bins, histtype='bar', rwidth=10)
plt.show()
```

```
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [6, 9, 3, 4, 8, 6, 4, 8, 5]

plt.scatter(x, y, label='skitscat', color='k', marker='*', s=100)
plt.legend()
plt.show()
```

```
x = [1, 2, 3, 4, 5]
y1 = [1, 1, 2, 3, 5]
y2 = [0, 4, 2, 6, 8]
y3 = [1, 3, 5, 7, 9]

y = np.vstack([y1, y2, y3])

labels = ["Fibonacci ", "Evens", "Odds"]

fig, ax = plt.subplots()
ax.stackplot(x, y1, y2, y3, labels=labels)
ax.legend(loc='upper left')
plt.show()

fig, ax = plt.subplots()
ax.stackplot(x, y)
plt.show()
```

```
slices = [7, 2, 2, 13]
activities = ['sleeping', 'eating', 'working', 'playing']
cols = ['c', 'm', 'r', 'k']

plt.pie(slices, labels=activities, colors=cols, startangle=90,  shadow=True, 
        explode=(0, 0.1, 0, 0), autopct='%1.1f%%')
plt.show()
```
