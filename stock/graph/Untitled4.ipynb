```
import matplotlib.pyplot as plt
```

```
x = [1, 2, 3]
y = [5, 7, 4]

x2 = [1, 2, 3]
y2 = [10, 14, 12]

plt.plot(x, y, label='first line')
plt.plot(x2, y2, label='second line')
plt.xlabel('Flot number')
plt.ylabel('Important var')
plt.legend()
plt.title('Graph')
plt.show()
```

```
x = [2, 4, 6, 8, 10]
y = [6, 7, 8, 2, 4]

x2 = [1, 3, 5, 9, 10]
y2 = [7, 8, 2, 4, 2]
plt.bar(x, y, label='Bars1', color='r')
plt.bar(x2, y2, label='Bars2')
plt.legend()
plt.show()
```
